## 📥 Ingestão de Arquivo Excel

## 🎯 Objetivo Técnico

Realizar a ingestão de um arquivo no formato **.xlsx** utilizando a biblioteca `pandas`, carregando os dados em um **DataFrame** para validação estrutural e preparação para processamento posterior.

## ⚙️ Procedimento

- Importação da biblioteca `pandas`
- Definição do caminho do arquivo
- Leitura via `pd.read_excel()`
- Validação da carga com:
  - `df.shape`
  - `df.head()`

## 📊 Validações

- Verificação da quantidade de linhas e colunas
- Confirmação de leitura sem erro
- Disponibilização dos dados em memória

## 📌 Resultado

DataFrame carregado com sucesso e pronto para etapas subsequentes de tratamento, transformação ou análise.

```python
import pandas as pd

path = r"C:\AZURE_DEVOS_REPOS\RSM_DATALAKE_JOURNAL_ENTRIES\033-DTLK-Data Analytics_Viação Normandy do Triângulo Ltda_31.12.2025\bronze\balancete\balancete.xlsx"

df = pd.read_excel(path)

print("✅ Arquivo carregado com sucesso")
print(f"Linhas: {df.shape[0]}")
print(f"Colunas: {df.shape[1]}")

df.head()
```

## 📊 Inspeção de Estrutura do DataFrame

Criação de um DataFrame auxiliar contendo:

- Nome das colunas (`df.columns`)
- Tipo de dado correspondente no pandas (`df.dtypes`)

Objetivo: visualizar rapidamente o schema do dataset para validação estrutural.

```python
df_info = pd.DataFrame({
    "coluna": df.columns,
    "tipo_pandas": df.dtypes.astype(str)
})

df_info
```

## 📊 Análise de Valores Nulos

Criação de um DataFrame auxiliar contendo:

- Total de linhas do dataset
- Quantidade de valores nulos por coluna
- Percentual de nulos (%)

Objetivo: avaliar completude dos dados e identificar possíveis riscos para as etapas de tratamento e transformação.

```python
df_nulos = pd.DataFrame({
    "coluna": df.columns,
    "linhas_totais": len(df),
    "nulos_qtd": df.isna().sum(),
    "nulos_pct": (df.isna().mean() * 100).round(2)
})

df_nulos
```
