```python
# ============================================
# CÉLULA 4: ANÁLISE OBJETIVA + RELATÓRIO FINAL
# ============================================

import json
import pickle
import hashlib
from pathlib import Path

# Criar pasta de cache
CACHE_DIR = Path("cache_analises_comentario")
CACHE_DIR.mkdir(exist_ok=True)

PROMPT_RESUMO_COMENTARIO = """
Você é um professor especialista em concursos públicos.

Sua tarefa é transformar o comentário do professor em uma explicação mais clara, objetiva e fácil de revisar.

REGRAS IMPORTANTES:
- Use SOMENTE as informações presentes no comentário do professor
- Não invente fundamento jurídico
- Não acrescente jurisprudência, artigos ou doutrina que não estejam no comentário
- Seja direto, didático e resumido
- Foque no que realmente ajuda o aluno a entender o motivo da resposta

Estruture a resposta exatamente assim:

TEMA:
[tema principal em 1 linha]

EXPLICAÇÃO OBJETIVA:
[reescreva o comentário de forma mais clara e curta]

MOTIVO DA RESPOSTA CORRETA:
[explique em poucas linhas]

RESUMO PARA MEMORIZAÇÃO:
[escreva a regra principal em 1 frase]

DADOS DA QUESTÃO:

Enunciado:
{enunciado}

Resposta correta:
{resposta_correta}

Comentário do professor:
{comentario}
"""


# ============================================
# GERAR ID DE CACHE
# ============================================

def gerar_id_cache(questao_dict):

    base = "||".join([
        str(questao_dict.get("Enunciado", "")),
        str(questao_dict.get("Resposta_Correta", "")),
        str(questao_dict.get("Comentario_Professor", ""))
    ])

    return hashlib.md5(base.encode("utf-8")).hexdigest()


# ============================================
# VALIDAR QUALIDADE DA RESPOSTA
# ============================================

def validar_qualidade_analise(analise):

    score = 0
    texto = analise.strip()

    tamanho = len(texto)

    if 300 <= tamanho <= 1800:
        score += 30
    elif 200 <= tamanho <= 2200:
        score += 20
    elif tamanho > 100:
        score += 10

    estrutura = [
        "TEMA:",
        "EXPLICAÇÃO OBJETIVA:",
        "MOTIVO DA RESPOSTA CORRETA:",
        "RESUMO PARA MEMORIZAÇÃO:"
    ]

    score += sum(10 for item in estrutura if item in texto)

    if "\n" in texto:
        score += 10

    termos_ruins = ["não foi possível", "informação insuficiente", "erro", "n/a"]

    if not any(t in texto.lower() for t in termos_ruins):
        score += 10

    if "RESUMO PARA MEMORIZAÇÃO:" in texto and len(texto.split()) >= 40:
        score += 10

    return min(score, 100)


# ============================================
# GERAR RESUMO IA
# ============================================

def analisar_questao_com_ia_otimizada(questao_dict, max_tentativas=2):

    comentario = str(questao_dict.get("Comentario_Professor", "")).strip()

    if not comentario or comentario.lower() in ["n/a", "nan", "none"]:

        return """TEMA:
Não identificado

EXPLICAÇÃO OBJETIVA:
Não há comentário do professor disponível para resumir.

MOTIVO DA RESPOSTA CORRETA:
Não foi possível gerar a explicação sem o comentário.

RESUMO PARA MEMORIZAÇÃO:
Questão sem comentário do professor.
"""

    questao_id = gerar_id_cache(questao_dict)

    cache_file = CACHE_DIR / f"{questao_id}.pkl"

    if cache_file.exists():

        with open(cache_file, "rb") as f:
            return pickle.load(f)

    ultima_analise = None

    for tentativa in range(1, max_tentativas + 1):

        try:

            prompt_final = PROMPT_RESUMO_COMENTARIO.format(
                enunciado=questao_dict.get("Enunciado", "N/A"),
                resposta_correta=questao_dict.get("Resposta_Correta", "N/A"),
                comentario=comentario
            )

            config = {

                "anthropic_version": "bedrock-2023-05-31",
                "max_tokens": 1200,
                "temperature": 0.2,

                "system": """Você é um professor especialista em concursos.
Resuma comentários de professores de forma clara e objetiva.
Nunca invente informações.
Seja fiel ao comentário original.
""",

                "messages": [
                    {
                        "role": "user",
                        "content": prompt_final
                    }
                ]
            }

            response = bedrock_client.invoke_model(
                body=json.dumps(config),
                modelId=MODEL_ID,
                accept="application/json",
                contentType="application/json"
            )

            resposta = json.loads(response["body"].read().decode("utf-8"))

            analise = resposta.get("content", [{}])[0].get("text", "").strip()

            ultima_analise = analise

            score = validar_qualidade_analise(analise)

            print(f"📊 Score qualidade: {score}")

            if score >= 60:

                with open(cache_file, "wb") as f:
                    pickle.dump(analise, f)

                return analise

        except Exception as e:

            print("Erro:", e)

    if ultima_analise:

        with open(cache_file, "wb") as f:
            pickle.dump(ultima_analise, f)

        return ultima_analise

    return "Falha ao gerar análise."


# ============================================
# MONTAR RELATÓRIO COMPLETO DA QUESTÃO
# ============================================

def montar_relatorio_questao(questao_dict):

    enunciado = questao_dict.get("Enunciado", "N/A")

    alternativas = questao_dict.get("Alternativas", "N/A")

    resposta = questao_dict.get("Resposta_Correta", "N/A")

    comentario = questao_dict.get("Comentario_Professor", "N/A")

    resumo_ia = analisar_questao_com_ia_otimizada(questao_dict)

    relatorio = f"""

==================================================
ANÁLISE DE QUESTÃO
==================================================

ENUNCIADO
{enunciado}

ALTERNATIVAS
{alternativas}

RESPOSTA CORRETA
{resposta}

COMENTÁRIO DO PROFESSOR
{comentario}

--------------------------------------------------
RESUMO GERADO PELA IA
--------------------------------------------------

{resumo_ia}

==================================================

"""

    return relatorio


print("✅ Sistema de análise criado com sucesso!")
print("Agora o relatório inclui:")
print("✔ Enunciado")
print("✔ Alternativas")
print("✔ Comentário do professor")
print("✔ Resumo da IA")
print("✔ Cache automático")
print("✔ Validação de qualidade")
```

```python
# ============================================
# CÉLULA 5: GERAR PDF COM TODAS AS QUESTÕES
# ============================================

from datetime import datetime
from xhtml2pdf import pisa


def gerar_pdf_compacto(dataframe, caminho_pdf="apostila_questoes_teste.pdf"):

    def escape_html(text):

        if text is None:
            return "N/A"

        return (str(text)
                .replace("&","&amp;")
                .replace("<","&lt;")
                .replace(">","&gt;"))

    html_questoes = ""

    for i, row in dataframe.iterrows():

        q = row.to_dict()

        enunciado = escape_html(q.get("Enunciado","N/A"))

        gabarito = str(q.get("Resposta_Correta","")).upper()

        comentario = escape_html(q.get("Comentario_Professor",""))

        analise = analisar_questao_com_ia_otimizada(q)

        analise = escape_html(analise).replace("\n","<br/>")

        alternativas_html = ""

        for letra in ["A","B","C","D","E"]:

            texto = (
                q.get(f"alt_{letra.lower()}") or
                q.get(f"Alternativa_{letra}") or
                ""
            )

            texto = escape_html(texto)

            if letra == gabarito:

                alternativas_html += f"""
                <div class="alt correta">
                <b>{letra})</b> {texto}
                </div>
                """

            else:

                alternativas_html += f"""
                <div class="alt">
                <b>{letra})</b> {texto}
                </div>
                """

        html_questoes += f"""

        <div class="questao">

            <div class="num">
            QUESTÃO {i+1}
            </div>

            <div class="enunciado">
            {enunciado}
            </div>

            <div class="alternativas">
            {alternativas_html}
            </div>

            <div class="analise">
            {analise}
            </div>

        </div>

        """

    html = f"""

    <html>

    <head>

    <meta charset="UTF-8"/>

    <style>

    body{{
        font-family: Arial;
        font-size:8px;
        line-height:1.2;
        margin:20px;
    }}

    .titulo{{
        text-align:center;
        font-size:14px;
        margin-bottom:10px;
        font-weight:bold;
    }}

    .questao{{
        margin-bottom:18px;
        page-break-inside:avoid;
    }}

    .num{{
        font-weight:bold;
        font-size:9px;
        margin-bottom:4px;
    }}

    .enunciado{{
        margin-bottom:4px;
    }}

    .alt{{
        margin-left:10px;
    }}

    .correta{{
        background:#e8ffe8;
    }}

    .analise{{
        margin-top:4px;
        color:#333;
    }}

    </style>

    </head>

    <body>

    <div class="titulo">
    CADERNO DE QUESTÕES COMENTADAS
    </div>

    {html_questoes}

    <br/>

    <div style="font-size:7px;text-align:center">
    Gerado automaticamente • {datetime.now().strftime("%d/%m/%Y %H:%M")}
    </div>

    </body>

    </html>

    """

    with open(caminho_pdf,"wb") as pdf:

        pisa.CreatePDF(html,dest=pdf)

    print("✅ PDF gerado:",caminho_pdf)
    
```

```python
# ============================================
# CÉLULA 7: TESTE COM 1 QUESTÃO (AJUSTADA)
# ============================================

import re
import pandas as pd


# ------------------------------------------------
# Função segura para pegar coluna
# ------------------------------------------------
def safe_get(row, coluna, default=""):

    if coluna not in row.index:
        return default

    valor = row[coluna]

    if pd.isna(valor):
        return default

    valor = str(valor).strip()

    if valor.lower() in ["nan", "none", ""]:
        return default

    return valor


# ------------------------------------------------
# Limpar nome de arquivo
# ------------------------------------------------
def limpar_nome_arquivo(nome):

    return re.sub(r"[^\w\-_.]", "_", str(nome))


# ------------------------------------------------
# Normalizar gabarito
# ------------------------------------------------
def normalizar_gabarito(valor):

    valor = str(valor)

    valor = valor.replace("Alternativa", "")
    valor = valor.replace(":", "")

    return valor.strip().upper()


# ------------------------------------------------
# Função principal de teste
# ------------------------------------------------
def testar_questao(dataframe, indice=0):

    print("🧪 TESTE - Processando 1 questão")
    print("=" * 70)

    try:

        if dataframe.empty:
            raise Exception("Dataset vazio")

        if indice >= len(dataframe):
            raise Exception("Índice fora do dataset")

        row = dataframe.iloc[indice]

        comentario = safe_get(row, "Comentario_Professor")

        # -------------------------------------
        # OBJETO QUESTÃO
        # -------------------------------------
        questao = {

            "Numero_Questao": safe_get(row, "Numero_Questao", indice+1),

            "Resultado": safe_get(row, "Resultado"),

            "Enunciado": safe_get(row, "Enunciado"),

            "Comentario_Professor": comentario,

            "Resposta_Correta": normalizar_gabarito(
                safe_get(row, "Resposta_Correta")
            ),

            "Alternativa_A": safe_get(row, "Alternativa_A"),
            "Alternativa_B": safe_get(row, "Alternativa_B"),
            "Alternativa_C": safe_get(row, "Alternativa_C"),
            "Alternativa_D": safe_get(row, "Alternativa_D"),
            "Alternativa_E": safe_get(row, "Alternativa_E")
        }

        # -------------------------------------
        # LOGS
        # -------------------------------------
        print(f"\n📋 Questão: {questao['Numero_Questao']}")
        print(f"Gabarito: {questao['Resposta_Correta']}")

        print("\n📖 Enunciado:")
        print(questao["Enunciado"][:200], "...")

        if comentario:
            print("\n🧠 Comentário professor: ✔ encontrado")
        else:
            print("\n⚠️ Comentário professor: não encontrado")

        print("\n🤖 Enviando para IA...")
        print("⏳ Aguarde...\n")

        # -------------------------------------
        # CHAMADA IA
        # -------------------------------------
        analise = analisar_questao_com_ia_otimizada(questao)

        if not analise:
            raise Exception("IA retornou resposta vazia")

        # -------------------------------------
        # PREVIEW IA
        # -------------------------------------
        print("\n📊 RESPOSTA IA (preview)")
        print("-" * 70)

        print(analise[:600])

        print("-" * 70)

        print(f"\n📏 Tamanho da resposta: {len(analise)} caracteres")

        # -------------------------------------
        # GERAR PDF
        # -------------------------------------
        print("\n📄 Gerando PDF...")

        nome_pdf = limpar_nome_arquivo(
            f"questao_{questao['Numero_Questao']}.pdf"
        )

        gerar_pdf_questao(
            questao_dict=questao,
            analise_ia=analise,
            output_path=nome_pdf
        )

        print("\n" + "=" * 70)
        print("✅ TESTE FINALIZADO")
        print(f"📁 PDF gerado: {nome_pdf}")
        print("=" * 70)

        return questao, analise

    except Exception as e:

        print("\n❌ ERRO\n")

        print(str(e))

        import traceback
        traceback.print_exc()

        return None, None


# ------------------------------------------------
# EXECUTAR TESTE
# ------------------------------------------------

print("🚀 Iniciando teste do pipeline...\n")

questao_teste, analise_teste = testar_questao(df, indice=0)
```

```python
# ============================================
# CÉLULA 8: PROCESSAR TODAS AS QUESTÕES
# ============================================

from pathlib import Path
import time


def processar_todas_questoes(dataframe, nome_pdf_final="apostila_questoes_teste.pdf"):

    print("🚀 PROCESSANDO TODAS AS QUESTÕES")
    print("=" * 70)

    if dataframe.empty:
        print("❌ DataFrame vazio")
        return

    total = len(dataframe)

    print(f"📊 Total de questões: {total}\n")

    questoes_processadas = []

    sucessos = 0
    erros = 0

    tempo_inicio = time.time()

    for index, row in dataframe.iterrows():

        try:

            numero = index + 1

            progresso = (numero / total) * 100

            print(f"📌 Questão {numero}/{total} ({progresso:.1f}%)")

            comentario = safe_get(row, "Comentario_Professor")

            questao = {

                "Numero_Questao": safe_get(row, "Numero_Questao", numero),

                "Resultado": safe_get(row, "Resultado"),

                "Enunciado": safe_get(row, "Enunciado"),

                "Comentario_Professor": comentario,

                "Tem_Comentario": comentario != "",

                "Resposta_Correta": normalizar_gabarito(
                    safe_get(row, "Resposta_Correta")
                ),

                "Alternativa_A": safe_get(row, "Alternativa_A"),
                "Alternativa_B": safe_get(row, "Alternativa_B"),
                "Alternativa_C": safe_get(row, "Alternativa_C"),
                "Alternativa_D": safe_get(row, "Alternativa_D"),
                "Alternativa_E": safe_get(row, "Alternativa_E")
            }

            # --------------------------------
            # CHAMAR IA
            # --------------------------------
            analise = analisar_questao_com_ia_otimizada(questao)

            if not analise:
                raise Exception("IA retornou resposta vazia")

            questao["Analise_IA"] = analise

            questoes_processadas.append(questao)

            sucessos += 1

        except Exception as e:

            print(f"❌ Erro na questão {numero}: {str(e)}")

            erros += 1

            continue

    # --------------------------------
    # GERAR PDF FINAL
    # --------------------------------
    print("\n📄 Gerando PDF final...")

    gerar_pdf_compacto(
        pd.DataFrame(questoes_processadas),
        nome_pdf_final
    )

    tempo_total = time.time() - tempo_inicio

    print("\n" + "=" * 70)

    print("🎉 PROCESSAMENTO FINALIZADO")

    print("=" * 70)

    print(f"✅ Sucesso: {sucessos}/{total}")
    print(f"❌ Erros: {erros}/{total}")

    print(f"\n📄 PDF final: {nome_pdf_final}")

    print(f"⏱ Tempo total: {tempo_total:.1f}s")
```

```python
# ============================================
# GERAR PDF DE REVISÃO ULTRA COMPACTA
# ============================================

import pandas as pd
import json
from pathlib import Path
from datetime import datetime
from xhtml2pdf import pisa

CAMINHO_DATASET = Path(r"C:\welligton-pos-IA\rpa-ia\simulado_const_1.xlsx")
PDF_SAIDA = "revisao_flashcards_teste1.pdf"


PROMPT_REVISAO_CURTA = """
Você é um professor especialista em concursos públicos.

Baseado EXCLUSIVAMENTE no comentário do professor, gere uma revisão extremamente curta.

REGRAS:

- Não invente conteúdo
- Use somente o comentário
- Seja extremamente direto

Formato da resposta:

REVISÃO:
[1 frase técnica que explica o conceito cobrado]

PALAVRAS-CHAVE:
[3 a 6 palavras-chave do tema]
"""


def gerar_revisao_curta(comentario):

    if not comentario or str(comentario).strip()=="":
        return "Sem comentário disponível."

    prompt = PROMPT_REVISAO_CURTA + "\n\nComentário:\n" + str(comentario)

    config = {
        "anthropic_version":"bedrock-2023-05-31",
        "max_tokens":200,
        "temperature":0.2,
        "messages":[{"role":"user","content":prompt}]
    }

    try:

        response = bedrock_client.invoke_model(
            body=json.dumps(config),
            modelId=MODEL_ID,
            accept="application/json",
            contentType="application/json"
        )

        resposta = json.loads(response["body"].read().decode())

        texto = resposta.get("content",[{}])[0].get("text","")

        return texto.strip()

    except Exception as e:

        return f"Erro: {str(e)}"


print("📂 Carregando dataset...")

df = pd.read_excel(CAMINHO_DATASET)

print("Total de questões:",len(df))


# ------------------------------------------------
# FILTRAR ERRADAS
# ------------------------------------------------

df_erradas = df[df["Resultado"].astype(str).str.lower().isin(["errada","incorreta"])]

print("Questões erradas:",len(df_erradas))


html_revisoes = ""

for i,row in df_erradas.iterrows():

    comentario = row.get("Comentario_Professor","")

    revisao = gerar_revisao_curta(comentario)

    revisao = revisao.replace("\n","<br/>")

    numero = row.get("Numero_Questao", i+1)

    html_revisoes += f"""

    <div class="card">

    <div class="numero">
    QUESTÃO {numero}
    </div>

    <div class="conteudo">
    {revisao}
    </div>

    </div>

    """


html = f"""

<html>

<head>

<meta charset="UTF-8"/>

<style>

body{{
font-family:Arial;
font-size:9px;
margin:20px;
}}

.titulo{{
text-align:center;
font-size:16px;
font-weight:bold;
margin-bottom:20px;
}}

.card{{
margin-bottom:12px;
padding:6px;
border-bottom:1px solid #ddd;
}}

.numero{{
font-weight:bold;
font-size:10px;
margin-bottom:2px;
}}

.conteudo{{
color:#333;
}}

</style>

</head>

<body>

<div class="titulo">
REVISÃO RÁPIDA DE QUESTÕES<br/>
{datetime.now().strftime("%d/%m/%Y")}
</div>

{html_revisoes}

</body>

</html>

"""


print("\n📄 Gerando PDF...")

with open(PDF_SAIDA,"wb") as pdf:

    pisa.CreatePDF(html,dest=pdf)


print("\n✅ PDF gerado!")

print("Arquivo:",PDF_SAIDA)
```
